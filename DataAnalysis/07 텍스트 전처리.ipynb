## 1. 텍스트 클렌징

텍스트 분석이전에 분석에 방해가 되는 요소를 제거하여 분석에 필요한 텍스트만 남기는 작업 입니다.   
특수문자를 제거하는 작업과 분석에 방해가 되는 특정 문자를 제거하는 불용어 제거 등이 포함됩니다.

```python
text = df.loc[0,'text']
text
```

### 1.1 특수문자 제거하기
특수 문자 제거는 파이썬의 내장 모듈인 `re` 모듈을 import하여 `sub()` 함수를 이용해 줍니다.
> re.sub(수정될 문자열의 정규식, 수정후의 문자열, 문자열)

```python
## 특수문자 정규식을 첫번째 인자로 입력
## 공백을 두번째 인자로 입력
import re

rtext = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》\xa0\n\d+]', '', text)
rtext
```

### 1.2 불용어 제거하기
불용어 제거는 파이썬의 문자열 함수와 리스트, 반복문등을 이용하여 구성 가능합니다.    
문자열의 `split()`,`replace()`,`join()`등의 함수를 활용해줍니다.

```python
## 먼저 단어를 띄어쓰기 기준으로 스플릿
sptext = rtext.split(' ')
sptext
```

단어로 스플릿 이후에는 반복문을 통해서 해당 단어가 불용어에 속하는지 `if` `in` 문을 활용하여 체크하고 속하지 않으면 새로운 리스트에 담아 join해줍니다.

```python
nw = []
for w in sptext:
    if w not in ['','초','이','등']:
        nw.append(w)
rtext = ' '.join(nw)
rtext
```

#### 전체 데이터에 적용하기

```python
import re

text_sub = []
for text in df['text']:
    rtext = str(text).strip()
    rtext = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》\xa0\n\d+]', '', rtext)
    sptext = rtext.split(' ')
    nw = [w for w in sptext if w not in ['','초','이','등']]
#     nw = []
#     for w in sptext:
#         if w not in ['','초','이','등']:
#             nw.append(w)
#     rtext = ' '.join(nw)
    rtext = ' '.join(nw)
    text_sub.append(rtext)
df['text_sub'] = text_sub
df
```

## 2. 토큰화

텍스트를 분석하기 위해서는 텍스트의 특정 단위가 필요하고 이런 단위을 토큰이라고 합니다.   
기본적으로 **토큰화**라고 하면 특정 문자들을 토큰으로 하여 자르는 과정이 됩니다.   

### 2.1 문자로 토큰화

문자로 토큰화 하는 방법중 가장 기본적인 방법은 `split()` 함수를 이용하는 겁니다.   
기본적으로 띄어쓰기 기준으로 스플릿하여 토큰화를 이룰 수 있습니다.

### 2.2 숫자 인코딩

문자 토큰을 숫치형으로 나타내는 가장 기본적인 방법은 문자 토큰에 **고유숫자**를 부여 하는 방법입니다.   
이렇게 숫자 토큰으로 변환하는 과정을 **인코딩**, 역으로 변환하는 과정을 디코딩이라 합니다.   
기본적인 인코딩 방식은 모든 문자 토큰을 찾아 고유 번호를 맵핑시킨 **딕셔너리**를 만드는 방식입니다.

```python
## 고유 토큰을 찾기위해 set 구조 활용
uq = set()
for text in df['text_sub']:
    sp = text.split(' ')
    for w in sp:
        uq.add(w)
uq
```

```python
## 고유 토큰에 고유 번호를 맵핑시킨 딕셔너리 구성
enc_dict = {}
for i,t in enumerate(uq):
    enc_dict[t] = i
enc_dict
```

```python
## 딕셔너리를 활용하여 인코딩
text = df.loc[0,'text_sub']
str_tk = text.split(' ')
enc_tk = [enc_dict[w] for w in str_tk]
enc_tk
```

```python
import pandas as pd
from mecab import MeCab
from kiwipiepy import Kiwi
import re


df = pd.read_csv('./data/reviews.csv')
mecab = MeCab()
kw = Kiwi()
```

```python
text = df.loc[0,'text']
text
```

#### mecab 모델을 활용하여 tag정보를 포함한 토큰화

```python
## 원하는 tag만 가져오기
use_tags = ['NNP', 'NNG', 'NP', 'IC', 'VV', 'MAG']

# r = []
# for w in kwords:
#     if w[1] in use_tags:
#         r.append(w[0])
# r
kwords = [w[0].lower() for w in kwords if w[1] in use_tags]
rnew_text = ' '.join(kwords)
rnew_text
```

#### kiwipie 모델을 활용하여 tag정보를 포함한 토큰화 

```python
# 객체 형태로 토큰과 테그가 묶임
kwords = kw.tokenize(text)
kwords
```

```python
## 원하는 테그 가져오기
use_tags = ['NNP', 'NNG', 'NP', 'IC', 'VV', 'MAG']
kwords = [w.form for w in kwords if w.tag in use_tags] 
rnew_text = ' '.join(kwords)
rnew_text
```

#### mecab 과 kiwipie의 테그 결과 비교

```python
# 판다스 출력 옵션 설정(무한대 출력)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

def show_tag(text):
    mc_text = mecab.pos(text)
    kw_text = [(t.form, t.tag) for t in kw.tokenize(text)] 
    max_len = max([len(mc_text),len(kw_text)])

    mc_text += [None for i in range(max_len - len(mc_text))]
    kw_text += [None for i in range(max_len - len(kw_text))]

    text_set = {
        'Mecab':mc_text,
        'Kiwi':kw_text
    }
    return pd.DataFrame.from_dict(text_set)
```

#### 전체데이터에 적용

## 4. 바이트페어 변환

형태소 분석의 단점중 하나는 토큰의 기준이 이미 정해져 있다는 것입니다.    
바이트페어 변환은 텍스트를 **바이트단위(문자)** 로 묶어가며 빈도 확률에 따라 최적화된 토큰셋을 만들어 냅니다.    
따라서 형태소 분석과 다르게 가지고 있는 텍스트뭉치에 따라 토큰이 최적화 되어 나온다는 장점이 있습니다.    

### 4.1 spm 모듈 활용
`sentencepiece` 모듈은 바이트페어 인코딩 기능이 있는 대표적인 파이썬 라이브러리입니다.   
sentencepiece 모듈을 활용하여 다음과정을 통해 토큰화가 가능합니다.
1. sentencepiece 모듈에서 SentencePieceTrainer 객체를 통해 텍스트뭉치에서 토큰셋(model) 파일을 만들어냅니다.   
2. sentencepiece 모듈의 SentencePieceProcessor객체에 생성한 토큰셋 파일을 변환기(인코더)를 만들어 냅니다.
3. 만든 인코더에서 encode() 함수를 활용하여 텍스트를 토큰화 합니다.

[sentencepiece](https://github.com/google/sentencepiece)

#### 텍스트 뭉치 데이터를 텍스트 파일로 생성

```python
import sentencepiece as spm
import pandas as pd
import re

df = pd.read_csv('./data/sentence2_test.csv')

# 추가 쓰기모드('a')로 텍스트 파일 열기
with open('./data/sentence2_test.txt', 'a', encoding='utf-8') as af:    
    for text in df['text']:
        text = str(text).strip()
        text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》\xa0\n\d+]', '', text)
        try:
            # kwords = mecab.morphs(text)
            # rnew_text = ' '.join(kwords)
            af.write(text+'\n')
        except:
                pass
```

#### 텍스트 파일로 부터 토큰 사전 및 sentencepiece 모델 생성
`SentencePieceTrainer`의 `train()` 함수를 이용하여 sentencepiece 모델을 생성할 수 있습니다.

```python
spm.SentencePieceTrainer.train(input='./data/sentence2_test.txt', # 텍스트 뭉치 파일
                            model_prefix='./model/spm3000',       # 출력 모델 파일 이름
                            vocab_size=3000                       # 토큰 개수
                            )
```

#### sentencepiece 모델을 활용하여 토큰화
sentencepiece 모델이 만들어진 후 `SentencePieceProcessor` 객체를 통해 생성된 모델을 다시 프로그램에 로드를 합니다. 이후 `encode()` 함수를 사용하여 토큰화 합니다.

```python
import sentencepiece as spm
import pandas as pd
import re

sp = spm.SentencePieceProcessor(model_file='./model/spm3000.model') # 모델 로드
df = pd.read_csv('./data/sentence2_test.csv')
text = df.loc[0,'text']
sp_list = sp.encode(text, out_type=int)
print(text)
print(sp_list)
```

```python
import sentencepiece as spm
import pandas as pd
import re

sp = spm.SentencePieceProcessor(model_file='./model/spm3000.model') # 모델 로드
df = pd.read_csv('./data/sentence2_test.csv')

def text_spm(text):
    text = str(text).strip()
    text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》\xa0\n\d+]', '', text)
    sp_list = sp.encode(text, out_type=str) # 문장을 토큰화
    sp_text = ' '.join(sp_list)
    return sp_text

df['text_spm'] = df['text'].apply(text_spm)  
df
```

> spm를 활용한 바이트페어 토큰화 결과중 `_`는 기존 문장의 띄어쓰기를 의미합니다. 따라서 띄어쓰기로 join 하여도 기존 문장의 띄어쓰기 정보를 잃어버리지 않습니다. 

## 5.  벡터화

고유 숫자로 인코딩하는 방식과 비슷한 방법으로 one-hot인코딩 방식이 있습니다.  
> `one-hot` 인코딩: 인코딩 숫자의 최댓값 만큼의 길이로 나타나는 1차원 배열에 토큰에 해당하는 인덱스만 1로 나타내고 나머지를 0으로 나타내는 방식 

이러한 one-hot 인코딩은 배열의 길이가 너무 길어지고 효율성이 떨어져 결과적으로 임베딩 기법을 사용하게 됩니다.   
임베딩 기법을 보기전에 먼저 다른 토큰화방식과 벡터화과정을 보도록 하겠습니다.

### 5.1 개수 벡터화

개수 벡터 특정 문장내에서 특정 토큰의 빈도수로 문장을 표현하는 벡터입니다.   
개수 벡터화는 `sklearn` 모듈의 `CountVectorizer` 모델을 이용하여 간단하게 구현할 수 있습니다.
[sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html)

```python
from sklearn.feature_extraction.text import CountVectorizer 
import pandas as pd

df = pd.read_csv('./data/sentence2_tag.csv').dropna()

## max_df : 해당 확률 이상으로 빈도가 많은 토큰을 제외
## min_df: 해당 확률 이하로 빈도가 적은 토큰을 제외
cvec = CountVectorizer(max_df=0.9, min_df=10)
cvec.fit(df['text_tag']) # 토큰생성 (띄어쓰기 기반)
names = cvec.vocabulary_ # 토큰셋 확인
print(len(names))
print(names)
```

### 5.2 TF-IDF 벡터화

TF-IDF라는 수치는 단어의 중요도를 나타내는 수치중 하나입니다.   
단어의 빈도인 TF에 단어의 역 문서 빈도인 IDF를 곱한값 입니다.    
이런 TF-IDF 값으로 문장을 벡터로 나타내는것이 TF-IDF 벡터화 입니다.


TF-IDF 벡터는 `sklearn` 모듈의 `TfidfVectorizer` 모델을 이용하여 간단하게 구현할 수 있습니다.   
[sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html#sklearn.feature_extraction.text.TfidfVectorizer)

```python
from sklearn.feature_extraction.text import TfidfVectorizer #tfid

tfvec = TfidfVectorizer(max_df=0.9, min_df=10)
tfvec.fit(df['text_tag']) # 토큰생성 (띄어쓰기 기반)
names = tfvec.vocabulary_ # 토큰셋 확인
print(len(names))
print(names)
```

```python
X = tfvec.transform(df.loc[:5,'text_tag']) # 인코딩
x = X.toarray()
print(x.shape)
print(x[0])
```

## 원하는 요소 찾기

### By 요소 선택자

1. `CSS_SELECTOR` : 태그이름과 클래스이름을 조합한 HTML 요소 경로 
> CSS_SELECTOR는 `태그이름.클래스이름(띄어쓰기를 .으로 치환)` 형태로 줄임표현이 가능  
줄임표현이 안될때는 전체 경로를 입력해야함
<img src="./img/20231002_164219.png" width = 500></img>
2. `XPATH` : 태그이름의 계층적 구조로 표현된 HTML 요소 경로  
 

### find_element() 함수를 활용하여 원하는 HTML 요소 찾기
1. By 클래스의 HTML요소 선택자를 find_element()의 첫번째 인자로 설정
2. 웹드라이브 상에서 원하는 요소를 오른쪽 클릭하여 검사(같은요소에 두번반복)
3. 웹 소스에서 설정한 선택자 방식의 경로를 복사하여 find_element()의 두번째 인자로 입력
4. find_element() 함수의 결과에서 원하는 작업 함수 사용

```python
from selenium.webdriver.common.by import By

## XPATH로 가져오기
## 단일 요소를 가져오기 좋음
xp = '//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[1]/div/div[2]/div/div[1]/div[1]/c-wiz[4]/section/div/div[2]/div[5]/div/div/button/span'
rb = driver.find_element(By.XPATH, xp)
rb.click()
```

```python
from selenium.webdriver.common.by import By

## CSS_SELECTOR로 가져오기
## 중복되는 값으로 인해 단일 요소를 찾기 힘듬
rb = driver.find_element(By.CSS_SELECTOR, 'span.VfPpkd-vQzf8d')
rb.click()
```

#### 드래그하여 더많은 리뷰 노출시키기
> `execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', 드래그가있는 요소)` 활용하여 드래그  
브라우저 창에있는 드래그는 CSS_SELECTOR 선택자가 html인 요소를 활용 

```python
from time import sleep
import random

xp = '//*[@id="yDmH0d"]/div[5]/div[2]/div/div/div/div/div[2]'
# xp = '//*[@id="yDmH0d"]/div[4]/div[2]/div/div/div/div/div[2]'
review_box = driver.find_element(By.XPATH, xp) # 드래그가 있는 요소 가져오기
for i in range(10):
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', review_box )
    sleep(random.randint(2, 3)) # 2~3초 쉬었다가 반복
    
```

#### 여러개의 요소를 한번에 찾아오기
> `find_elements()` 함수에 같은 경로를 가진 요소 선택자를 넣어 여러개의 요소를 리스트로 가져옴   

<img src="./img/20231002_164557.png" width = 300 align = "left"></img>

```python
## CSS_SELECTOR
## CSS_SELECTOR는 중복으로 활용이 되기에 여러 요소를 찾아오기 좋음
reviews = driver.find_elements(By.CSS_SELECTOR, 'div.RHo1pe')
reviews
```

#### 요소에서 원하는 데이터 추출
> `get_attribute()` 함수를 활용하여 태그 안에있는 변수를 인자로 입력하여 데이터를 추출   
텍스트를 추출할때는 `'innerText'` 인자를 활용

<img src="./img/20231002_165404.png" width = 400 align = "left"></img>

```python
reviews[0].find_element(By.CSS_SELECTOR, 'div.h3YV2d').get_attribute('innerText')
```

```python
## data-review-id 테그의 데이터 추출
reviews[0].find_element(By.CSS_SELECTOR, 'header.c1bOId').get_attribute('data-review-id')
```

```python
## <테그> 텍스트 </> 데이터 추출
texts = []
for review in reviews:
    text = review.find_element(By.CSS_SELECTOR, 'div.h3YV2d').get_attribute('innerText')
    texts.append(text)
texts
```

#### 함수와 반복문을 활용하여 모든 요소의 데이터 추출

```python
def get_content(review):
    condic = {}
    condic['text'] = review.find_element(By.CSS_SELECTOR, 'div.h3YV2d').get_attribute('innerText').strip()
    condic['id'] = review.find_element(By.CSS_SELECTOR, 'header.c1bOId').get_attribute('data-review-id')
    condic['rat'] = len(review.find_elements(By.CSS_SELECTOR, 'span.Z1Dz7b'))
    condic['date'] = review.find_element(By.CSS_SELECTOR, 'span.bp9Aid').get_attribute('innerText').strip()
    return condic
```

```python
import pandas as pd
result = []

for content in driver.find_elements(By.CSS_SELECTOR, 'div.RHo1pe'):
    try:
        dic = get_content(content)
        result.append(dic)
    except:
        print('error')
        pass
    
df = pd.DataFrame(result)
df
```

```python
# 데이터 폴더 생성 후
import os 
os.makedirs('data', exist_ok=True)  # 디렉토리 생성 (이미 존재하면 무시

df.to_csv('./data/reviews.csv', index=False)
```

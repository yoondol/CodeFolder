```python
#1번문제
def extract_birthdate(sentence):
    # 문자열을 띄어쓰기로 나누기
    words = sentence.split()

    # 연, 월, 일을 저장할 변수
    year = ""
    month = ""
    day = ""

    for word in words:
        # 이미 생년월일이 추출된 경우 루프 종료
        if year and month and day:
            break

        # '2001년'처럼 '년'이 포함된 단어에서 연도 추출
        if '년' in word:
            year = word.replace('년', '')
        # '10월'처럼 '월'이 포함된 단어에서 월 추출
        elif '월' in word:
            month = word.replace('월', '').zfill(2) #2자리
        # '6일에'처럼 '일'이 포함된 단어에서 일 추출
        elif '일' in word:
            day = word.replace('일에', '').zfill(2) #2자리
        # '2001/10/06'처럼 슬래시로 구분된 날짜 형식
        elif '/' in word :
            date_parts = word.split('/')
            if len(date_parts) == 3:
                year, month, day = date_parts[0], date_parts[1], date_parts[2]
        # '20011006'처럼 붙어있는 형식
        elif len(word) >= 8 and word[:8].isdigit():
            year, month, day = word[:4], word[4:6], word[6:8]

    # 모두 있을 경우 형식에 맞춰 반환
    if year and month and day:
        return f"{year}/{month}/{day}"
    else:
        return "생년월일을 찾을 수 없습니다."

# 실행 예시
sentence = input("문장을 입력하세요: ")
birthdate = extract_birthdate(sentence)
print(f"분석된 생년월일: {birthdate}")
```

```python
# 2번 문제
# 기본 전화번호부 딕셔너리
phonebook = {
    "홍길동": "010-4444-5555",
    "김중앙": "010-9191-8181",
    "심청": "010-3232-5454"
}

# 이름을 검색하는 함수
def search_phonebook(name):
    # 입력된 이름이 포함된 항목을 찾음
    results = {k: v for k, v in phonebook.items() if name in k}

    # 검색 결과가 있으면 출력
    if results:
        for k, v in results.items():
            print(f"{k}  {v}")
    else:
        print("찾을 수 없습니다.")

# 추가하는 함수
def add_phonebook():
    # 이름과 전화번호 입력받기
    new_name = input("이름은? ")
    new_number = input("전화번호는? ")

    # 새로운 항목 추가
    phonebook[new_name] = new_number
    print(f"{new_name} 전화번호가 추가되었습니다.")

while True:
    name = input("이름>> ")

    # 프로그램 종료
    if name == 'exit':
        print("프로그램을 종료합니다.")
        break

    # 전화번호 추가
    if name == 'add':
        add_phonebook()
    # 전화번호 검색
    else:
        search_phonebook(name)
```

```python
#3번 문제
# 천 단위로 ',' 삽입하는 간단한 함수 (99999 이하의 숫자만 처리)
def add_commas(number):
    num_str = str(number)  # 숫자를 문자열로 변환
    if len(num_str) > 3:
        return num_str[:-3] + ',' + num_str[-3:]  # 마지막 3자리 앞에 쉼표 추가
    else:
        return num_str

# 숫자를 한글로 나타내기 위한 딕셔너리
number_to_korean = {
    1: "일", 2: "이", 3: "삼", 4: "사", 5: "오",
    6: "육", 7: "칠", 8: "팔", 9: "구", 0: ""
}
units = ["", "십", "백", "천"]

# 사용자로부터 숫자 입력 받기
num = int(input("숫자는? "))

# 99999 이하의 숫자만 입력받도록 처리
if num > 99999:
    print("99999 이하의 숫자만 입력 가능합니다.")
else:
    # 천 단위로 , 표시 (간단한 버전)
    formatted_num = add_commas(num)

    # 숫자를 한글로 변환
    num_str = str(num)
    length = len(num_str)

    korean_result = ""
    for i, digit in enumerate(num_str):
        digit_int = int(digit)
        if digit_int != 0:  # 0은 생략
            # 만 단위 처리
            if length - i > 4:
                if digit_int == 1 and length - i == 5:  # 만의 자리에서 '일' 생략
                    korean_result += "만"
                else:
                    korean_result += number_to_korean[digit_int] + "만"
            else:
                korean_result += number_to_korean[digit_int] + units[length - i - 1]

    # 결과 출력
    print(f"{formatted_num}원")
    print(f"{korean_result}원")
```

```python
#4번 문제

def game():
    # 시작 단어와 기본 변수 설정
    current_word = "apple"
    score = 0
    used_words = set([current_word])
    print(f"시작 단어: {current_word}")

    # 10회 반복
    for _ in range(10):
        # 입력 받기
        next_word = input(f"단어({current_word}): ").strip().lower()

        # 조건 1: 3~6 글자 단어 확인
        if len(next_word) < 3 or len(next_word) > 6:
            print("단어의 길이가 맞지 않습니다.(-1)")
            score -= 1
        # 조건 2: 이미 나온 단어 확인
        elif next_word in used_words:
            print("이미 나온 단어입니다.(-1)")
            score -= 1
        # 조건 3: 끝말잇기 여부 확인
        elif current_word[-1] != next_word[0]:
            print("끝말잇기가 안됩니다.(-1)")
            score -= 1
        else:
            # 조건을 만족하면 점수 획득
            print(f"1점 획득! 현재 점수 {score + 1}점")
            score += 1
            used_words.add(next_word)  # 사용된 단어 목록에 추가
            current_word = next_word  # 현재 단어 업데이트

    print(f"게임 종료! 최종 점수는 {score}점입니다.")

# 게임 실행
game()
```

```python
# 5번 문제
# 사용자로부터 숫자 입력 받기
numbers = input("숫자? ")

# 입력된 숫자를 공백으로 분리하여 리스트로 변환
numbers_list = numbers.split()

# 정수와 실수를 저장할 리스트
integers = []
floats = []

# 숫자 분류 및 합계 계산
for num in numbers_list:
    if '.' in num:
        floats.append(float(num))  # 실수로 변환 후 실수 리스트에 추가
    else:
        integers.append(int(num))  # 정수로 변환 후 정수 리스트에 추가

# 결과 출력
print(f"정수 {integers}의 합계 = {sum(integers)}")
print(f"실수 {floats}의 합계 = {sum(floats)}")
```

```python
# 6번 문제
# 모음 리스트 정의 (대소문자 무시)
vowels = "aeiouAEIOU"

while True:
    # 사용자로부터 문장 입력 받기
    sentence = input("문장 ? (끝내려면 end) ")

    # 종료 조건 확인
    if sentence.lower() == "end":
        break

    # 모음 위치에 밑줄을 추가할 리스트 생성
    result = ""

    for char in sentence:
        if char in vowels:
            result += "_"  # 모음 위치에 밑줄 추가
        else:
            result += " "  # 다른 문자는 공백으로 처리

    # 원본 문장과 밑줄을 함께 출력
    print(f"문장 {sentence}")
    print(f"모음 {result}")
```
